```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# get data
df_preprocessed = pd.read_csv("Absenteeism_preprocessed.csv")
df_preprocessed.head(10)
```

```python
# Decide threshhold for target
df_preprocessed["Absenteeism Time in Hours"].median()
```

```python
# get encoded target values in 1-D array based on time below and above median
targets = np.where(df_preprocessed["Absenteeism Time in Hours"] > 
                   df_preprocessed["Absenteeism Time in Hours"].median(), 1, 0)
targets
```

```python
# add target to df
df_preprocessed["Excessive Absenteeism"] = targets
df_preprocessed.head()
```

```python
# check spread of targets - % of 1 values
targets.sum() / targets.shape[0]
```

## Checkpoint with Encoded Targets

```python
df_targets = df_preprocessed.drop(columns=["Absenteeism Time in Hours"])
df_targets.head()
```

```python
# check to see if dfs are different
df_targets is df_preprocessed
```

```python
# get features
unscaled_inputs = df_targets.iloc[:,:-1]
unscaled_inputs.head()
```

```python
# standardizing features
from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# fit_transform data
# scaled_inputs = sc.fit_transform(unscaled_inputs)
```

```python
unscaled_inputs.columns.values
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('scaler', StandardScaler(), ['Month Values',
       'Day of the Week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Children', 'Pets'])], remainder='passthrough')
```

```python
scaled_inputs = np.array(ct.fit_transform(unscaled_inputs))
scaled_inputs
```

```python
scaled_inputs[0]
```

```python
scaled_inputs
```

```python
# split training and testing data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size = 0.8, random_state = 20)
```

```python
# check training and testing shapes
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# fit training data
regressor = LogisticRegression()
regressor.fit(X_train, y_train)
```

```python
# get accuracy of model using training data
regressor.score(X_train, y_train)
```

```python
# get accuracy of model using testing data
# predictions
predictions = regressor.predict(X_test)
```

```python
metrics.accuracy_score(y_test, predictions)
```

## Summary Table

```python
# creating summary df with coefficients for each column
# get feature names for columns
features = ['Month Values',
       'Day of the Week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Children', 'Pets', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
```

```python
# create df
summary = pd.DataFrame(columns=['Features'], data=features)
summary["Coefficient"] = np.transpose(regressor.coef_)
summary
```

```python
summary.index = summary.index + 1
summary
```

```python
# create value at index 0
summary.loc[0] = ['Intercept', regressor.intercept_[0]]
summary.sort_index(inplace = True)
summary
```

```python
# get exponentials of coefficients
summary["Odds Ratio"] = np.exp(summary.Coefficient)
```

```python
summary
```

```python
# sort by greatest odds
summary.sort_values('Odds Ratio', ascending = False)
```

## Testing Data Accuracy

```python
regressor.score(X_test, y_test)
```

```python
# predicting the probability of target from features
predicted_proba = regressor.predict_proba(X_test)
predicted_proba
```
